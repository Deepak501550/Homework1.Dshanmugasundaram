```python
#q1 

import random
import matplotlib.pyplot as plt

def simulate_game():
   
    boxes = [10, 10, 100]
    random.shuffle(boxes)
    initial_choice = random.choice([0, 1, 2])

    if boxes[initial_choice] == 100:
        friend_choice = random.choice([i for i in range(3) if i != initial_choice])
    else:
        friend_choice = [i for i in range(3) if i != initial_choice and boxes[i] == 10][0]

    sticking_outcome = boxes[initial_choice]
    switching_outcome = boxes[[i for i in range(3) if i not in (initial_choice, friend_choice)][0]]

    return sticking_outcome, switching_outcome

def simulate_multiple_games(num_games):
    sticking_wins = 0
    switching_wins = 0

    for _ in range(num_games):
        sticking_outcome, switching_outcome = simulate_game()

        if sticking_outcome == 100:
            sticking_wins += 1
        if switching_outcome == 100:
            switching_wins += 1

    sticking_probability = sticking_wins / num_games
    switching_probability = switching_wins / num_games

    return sticking_probability, switching_probability

def create_probability_graph(sticking_probability, switching_probability):
    strategies = ['Stick', 'Switch']
    probabilities = [sticking_probability, switching_probability]

    colors = ['green', 'blue']  

    plt.bar(strategies, probabilities, color=colors)  
    plt.xlabel('Strategy')
    plt.ylabel('Probability')
    plt.title('Probability of Winning the $100 Bill')
    plt.ylim(0, 1)
    plt.show()

num_games = 10000



sticking_probability, switching_probability = simulate_multiple_games(num_games)

# Create the probability graph with different colors
create_probability_graph(sticking_probability, switching_probability)
print("\n")
print(" The value of the Sticking_Probability:",sticking_probability);
print("The value of the switching_Probability:",switching_probability)
```

```python
#Q2.

import random
import math
import numpy as np
import matplotlib.pyplot as plt

def monte_carlo_integral(f, a, b, num_samples):
    integral = 0
    for _ in range(num_samples):
        x = random.uniform(a, b)
        integral += f(x)

    integral /= num_samples
    return integral

def f(x):
    return x * np.cos(71 * x) + np.sin(13 * x)

def main():
    """
    Computes the integral of f(x) = x * cos(71x) + sin(13x) from 0 to 1 using Monte Carlo simulation.
    """

    a = 0
    b = 1
    num_samples = 10000

    integral = monte_carlo_integral(f, a, b, num_samples)

    print("The integral is:", integral)

    x_values = np.linspace(a, b, 100)
    y_values = f(x_values)

    plt.plot(x_values, y_values, color='red')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Plot of f(x) = x * cos(71x) + sin(13x)')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
